```python
import pandas as pd
df = pd.read_csv('/Users/suprita/supriML/fake_news_detector/fake_or_real_news.csv')
df
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
# TF-IDF will let us know what the most important terms in a document are
from sklearn.feature_extraction.text import TfidfVectorizer
# LinearSVC is good for text classification
from sklearn.svm import LinearSVC

# binary encoding for target
df['fake'] = df['label'].apply(lambda x: 0 if x == 'REAL' else 1)
del df['label']
df

 
```

```python
# 80-20 train and test split
x, y = df['text'], df['fake']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# vectorize article text
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7) # remove words like and, the, is, etc, and any word that appears in more than 70% of documents
x_train_vectorized = vectorizer.fit_transform(x_train)
x_test_vectorized = vectorizer.transform(x_test)

# train svc model
classifier = LinearSVC()
classifier.fit(x_train_vectorized, y_train)
```

```python
classifier.score(x_test_vectorized, y_test)
```
